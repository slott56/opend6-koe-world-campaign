# Temple of the Jackal Low-Level Priesthood

Lower-level priests, novices, and pilgrims.

```python
priest_text = """
    Name: Xorn Priests
    Characteristics:
    STR:13	DEX:16	CON:11	BODY:11	INT:10
    EGO:12	PRE:10	COM:10	PD:3	ED:2
    SPD:3	REC:5	END:22	STUN:23	Running:6
    Swimming:2
    Powers and Skills: Weapons Familiarity Group,Religion 11-,HKA 1/2d6 K,Strike 7d6,Block
    25+ Disadvantages:
    Fear Xorn: Common Situation, Strong Intensity, ; Jackal Follower: Uncommon Situation, Strong Intensity
"""

novice_text = """
    Name: Xorn Novices
    Characteristics:
    STR:13	DEX:16	CON:12	BODY:11	INT:10
    EGO:12	PRE:10	COM:10	PD:3	ED:2
    SPD:3	REC:5	END:24	STUN:24	Running:6
    Swimming:2
    Powers and Skills: Weapons Familiarity Group,Religion 11-,Survival 12-,Tracking 12-,Transport Familiarity 8-,Professional Skills 11-
    25+ Disadvantages:
    Fear Xorn: Common Situation, Strong Intensity, ; Jackal Follower: Uncommon Situation, Strong Intensity
"""

pilgrim_text = """
    Name: Xorn Pilgrims
    Characteristics:
    STR:13	DEX:13	CON:11	BODY:11	INT:10
    EGO:10	PRE:10	COM:10	PD:3	ED:2
    SPD:2	REC:5	END:22	STUN:23	Running:6
    Swimming:2
    Powers and Skills: Weapons Familiarity Group,Religion 9-,Survival 11-,Professional Skills 11-
    0+ Disadvantages:
    Fear Xorn: Common Situation, Strong Intensity, ; Jackal Follower: Uncommon Situation, Strong Intensity, ; Ability 9: ; Ability 10: ; Ability 11: ; Ability 12: ;
"""
```

```python
import hero
```

```python
from opend6_tools.character import *
```

```python
hero.opend6_seed(hero.HeroCharacter.from_text(priest_text))
```

```python
priest = Character(
    name='Devoted Xorn Priests',
    physique=Physique(3*D+2),
    agility=Agility(4*D, {'fighting': 1*D, 'melee combat': 2}),
    coordination=Coordination(4*D),
    intellect=Intellect(3*D+2),
    acumen=Acumen(3*D+2),
    charisma=Charisma(3*D+2, {'intimidation': 3*D+1}),
    body=22,
    move=10.0,
    special_abilities=[
        NaturalHandWeapon(2, "Martial arts skills: 1D+2 bare-hand attack"),
        NaturalArmor(2, "Martial Block"),
    ],
    disadvantages=[
        Devotion(1, "Great Priest Xorn"),
        Employed(2, "Following Xorn"),
    ]
)
priest.budget
```

```python
hero.opend6_seed(hero.HeroCharacter.from_text(novice_text))
```

```python
novice = Character(
    name='Devoted Xorn Novices',
    physique=Physique(3*D, {'stamina': 1*D}),
    agility=Agility(3*D, {'fighting': 1*D}),
    coordination=Coordination(3*D),
    intellect=Intellect(3*D),
    acumen=Acumen(3*D),
    charisma=Charisma(3*D),
    body=22,
    move=10.0,
    special_abilities=[
        NaturalHandWeapon(1, "Martial arts skills: 1D bare-hand attack"),
        NaturalArmor(1, "Martial Block"),
    ],
    disadvantages=[
        Devotion(1, "Great Priest Xorn"),
        Employed(2, "Following Xorn"),
    ]
)
novice.budget
```

```python
hero.opend6_seed(hero.HeroCharacter.from_text(pilgrim_text))
```

```python
pilgrim = Character(
    name='Temple of the Jackal Pilgrims',
    physique=Physique(2*D+2),
    agility=Agility(2*D+2, {'fighting': 0*D+2}),
    coordination=Coordination(2*D+2),
    intellect=Intellect(2*D+2),
    acumen=Acumen(2*D+2),
    charisma=Charisma(2*D+2, {'intimidation': 1*D}),
    body=22,
    move=10.0,
)
pilgrim.budget
```
