```python
highwaymen_text = """
    Name: Bandit Army Highwaymen
    Characteristics:
    STR:10	DEX:14	CON:10	BODY:11	INT:10
    EGO:12	PRE:8	COM:2	PD:2	ED:2
    SPD:3	REC:4	END:20	STUN:21	Running:6
    Swimming:2
    Powers and Skills: Weapons Familiarity Group,Riding 12-,Survival 11-
    0+ Disadvantages:
    Career Criminals: Less Powerful Capabilities, 11- Appearance, ; Greedy: Common Situation, Strong Intensity
"""
```

```python
import hero
```

```python
from opend6_tools.character import *
```

```python
hero.opend6_seed(hero.HeroCharacter.from_text(highwaymen_text))
```

```python
highwaymen = Character(
    name='Bandit Army Highwaymen',
    physique=Physique(3*D+2),
    agility=Agility(3*D+1, {'fighting': 2, 'melee combat': 2}),
    coordination=Coordination(3*D, {'marksmanship': 1}),
    intellect=Intellect(3*D, {'traps': 1}),
    acumen=Acumen(3*D, {'survival': 1, 'tracking': 1}),
    charisma=Charisma(2*D),
    body=22,
    move=10,
    disadvantages=[
        Infamy(2, "Known outlaws"),
        Quirk(2, "greedy"),
    ]
)
highwaymen.budget
```
