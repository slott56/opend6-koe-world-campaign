# Chapter 3 Grifters and Confidence Men

```python
grifter_text = """
        Name: Grifters
        Characteristics:
        STR:10	DEX:14	CON:10	BODY:10	INT:10
        EGO:11	PRE:8	COM:2	PD:2	ED:2
        SPD:3	REC:4	END:20	STUN:21	Running:6
        Swimming:2
        Powers and Skills: Streetwise 10-,Acting 10- (Confidence Schemes)
        0+ Disadvantages:
        Career Criminals: Less Powerful Capabilities, 11- Appearance, ; Greedy: Common Situation, Strong Intensity
"""
```

```python
import hero
```

```python
from opend6_tools.character import *
```

```python
hero.opend6_seed(hero.HeroCharacter.from_text(grifter_text))
```

```python
grifter = Character(
    name='Grifters',
    physique=Physique(3*D, ),
    agility=Agility(3*D, {'fighting': 1*D}),
    coordination=Coordination(3*D, {'sleight of hand': 1*D, 'throwing': 1*D}),
    intellect=Intellect(3*D, {'cultures': 1*D, 'speaking': 1*D}),
    acumen=Acumen(3*D+2, {'disguise': 1*D, 'gambling': 1*D, 'streetwise': 1*D}),
    charisma=Charisma(3*D+1, {'bluff': 1*D, 'charm': 1*D, 'persuasion': 1*D}),
    body=20,
    move=10,
    disadvantages=[
        Infamy(2, "known criminal"),
        Quirk(1, "greedy"),
    ]
)
grifter.budget
```
