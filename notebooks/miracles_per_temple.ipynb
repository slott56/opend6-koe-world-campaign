## Temple-specific difficulty offsets

Use a "placeholder" spell, one with essentially no difficulty, to compare the temples and assure they're more-or-less equitable.

```python
place_holder = Miracle(
    name="placeholder",
    notes="N/A",
    effect=Effect("None", 0),
    duration=DurationAspect("1 sec"),
    range=RangeAspect("1m"),
    casting_time=CastingTimeAspect("1 sec"),
    speed=SpeedAspect.based_on("range", "instant"),
)
for t in TEMPLES:
    specific = place_holder.finalize(other_aspects=t.aspects['other_aspects'])
    # print(t.name, specific.difficulty)
    print(t.name, specific.difficulty)
    # print(display(specific))
```

## Summarize difficulties

```python
summary = [
    ('Invocation',) + tuple(t.name for t in TEMPLES)
]
for invocation_name, invocation in INVOCATIONS.items():
    row = (invocation_name,)
    for temple in TEMPLES:
        # Create a specific Miracle from Template Miracle + Temple details.
        specific = invocation.finalize(other_aspects=temple.aspects['other_aspects'])
        row += (specific.difficulty,)
    summary.append(row)
summary
```

Locate spells with difficulties that are too low.

```python
for row in summary[1:]:
    if any(d < 5 for d in row[1:]):
        print(row)
```
