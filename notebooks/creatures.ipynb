3. Bright realm of sand and wind -- djinni and air elementals -- Loud Dragon
4. Dark realm smoke and ash -- giants, trolls, ogres
5. Dark realm of cold and mist -- water elementals -- Acid Dragon
6. Bright realm of fire -- fire elementals -- Fire Dragon

8. Ælvish realm of caves and earth -- dwarves, earth elementals -- Earth Dragon (aka Wyvern)
9. A hidden realm -- Metal Men.

## Ælvish realm of mountains and forests

Elf, Gnome, Goblin and Kobold are all -- essentially -- the same creature with distinct names.

There are two ælvish realms with humanoids. It is likely that inhabitants pass between the human realm and these other two.
This realm's inhabitants tend to be somewhat larger than humans.

```python
elf = Creature(
    name="Elf, Gnome, Goblin, Kobold",
    description="Inhabitants of the realm of mountains and forests",
    agility=Agility(3 * D, {'acrobatics': 2, 'dodge': 2, 'fighting': 2, "melee combat": 2, 'stealth': 2}),
    coordination=Coordination(3 * D, {"marksmanship": 2, "throwing": 1}),
    physique=Physique(3 * D,
                      {'running': 2, 'stamina': 2}),
    intellect=Intellect(3 * D, {"healing": 2, "scholar": 2, "traps": 2}),
    acumen=Acumen(3 * D, {"crafting": 2, "survival": 2}),
    charisma=Charisma(3 * D, {'command': 1, "intimidation": 2}),
    extranormal=Magic(3*D),
    disadvantages=[
        CulturalUnfamiliarity(2, "Lost in human company"),
    ],
    advantages=[
        Equipment(2, "Weapon or survival gear unknown to humans"),
    ],
    special_abilities=[
        CombatSense(2, "Surprise reduced by 4"),
    ],
    realm="Ælvish, mountains and forests",
)
print(display(elf))
```

## Compute a monster encounter probability table

The odds of seeing a monster are the inverse of the monster's overall dice budget. Powerful monsters are rare.

What's a bit tricky is pushing this into a context where a handful of D6 can select a monster from a complex, weighted collection of choices.

One choice is to slice up the distribution of nD6 into bands with width = P(monster).

Another choice is a cascading series of smaller tables relying the uniform distribution of D6 values.

```python
monsters = {
    name: thing
    for name, thing in globals().items()
    if isinstance(thing, Creature)
}
# monsters.keys()
```

We'll create the master list of creatures with a "probability" of the inverse of the creature's budget.

We'll find the creatures by examining all global variables (created in previous cells) with a value that's a `Creature` object.

```python
budgets = {
    name: (thing.budget['Attributes'] + thing.budget['Skills'] + thing.budget['Options']).n
    for name, thing in globals().items()
    if isinstance(thing, Creature)
}
```

```python
len(budgets)
```

```python
f"{min(budgets.values())} - {max(budgets.values())}"
```

```python
top = max(budgets.values())
ranks = {
    name: top-dice+1
    for name, dice in budgets.items()
}
```

This is lovely, but, the unknown realm is an exception, being both weak **and** rare.

We need to fiddle this slightly so the unknown realm creatures are tied for being most rare, that is, a rank of a lowly '1'.

```python
fixes = [
    name
    for name, thing in globals().items()
        if isinstance(thing, Creature) and thing.realm == 'unknown realm'
]
for nm in fixes:
    ranks[nm] = 1
```

## Map Monsters to nD6 frequencies

With a kind of flip-flop sort we can fiddle the monster list to put rare ones at each  end and common ones in the middle.
This create a normalish distribution.
The actual distibution based on budgets is somewhat flatter than this.

We do this by sort; extract the odd-index items in order;
reverse the odd-index items; extend the even-index-items with the reversed odd-index items.

```python
ranks_sorted = sorted((r, name) for name, r in ranks.items())
ranks_sorted_odd = ranks_sorted[1::2]
ranks_sorted_even = ranks_sorted[::2]
ranks_normalized = ranks_sorted_even
ranks_normalized.extend(reversed(ranks_sorted_odd))
ranks_normalized
```

We can map an ``nD6`` distribution (for an appropriate *n*) to these monsters, with a few lumps to put multiple dice rolls together to get (about) the right distribution.

The 19 monsters and 16 values in 3D6 is a near-miss that's only a coincidence. This requires 4D6 and some lumping.

When rolling $n\texttt{D6}$, there will be $6n-(n-1) = 5*n+1$ distinct values, $c$.
3 dice yields 16 values from 3 to 18.

Dice required is $n = \lceil \frac{c-1}{5} \rceil$.

```python
from math import ceil
n_dice = ceil((len(ranks_normalized)-1)/5)
n_dice
```

What is the distribution of nD6?
For $n$ dice of $s$ sides (6 in our case), the probability of getting the target value $T$ is this:

$$
P(n, s, T) = \sum\limits_{k=0}^{\lfloor \frac{T-n}{s} \rfloor}\bigl(-1\bigr)^k \binom{n}{k} \binom{T-sk-1}{n-1}
$$

We want to the probabilities for all values of $n \leq T < sn+1$.


```python
from math import comb, factorial
from functools import cache

@cache
def dice_prob(T, n, s=6):
    upper = (T - n) // s
    series = (
        (-1 if k % 2 == 1 else 1) * comb(n, k) * comb(T - s * k - 1, n - 1)
        for k in range(0, upper + 1)
    )
    return int(sum(series))

def distribution(n, s=6):
    return dict((t, dice_prob(t, n)) for t in range(n, n*6+1))
```

```python
def cumulative(dist_dict):
    total = 0
    for value, prob in dist_dict.items():
        total += prob
        yield total, value

nd6 = list(cumulative(distribution(n_dice)))
nd6
```

The following works by partitioning the 4D6 distribution into bins based on the monster ranking.

This isn't great because so many monsters have similar ranks.

```python
import bisect

bin_breaks = [n for n, value in nd6]
total = 0
prev = 1
for r, name in ranks_normalized:
    nd_rank = round(6**n_dice*r/ranks_total)
    total += nd_rank
    bin = bisect.bisect(bin_breaks, total)
    _, roll = nd6[bin]
    print(r, name, prev, roll)
    prev = roll
```

```python
extras = n_dice*5+1 - len(ranks_normalized)
extras
```

With some padding, we can simply map the monsters to the `nD6` distribution without knowing the actual frequencies.

```python
front_pad = extras // 2
back_pad = extras // 2 + extras % 2
ranks_padded = [(0, None)]*front_pad + ranks_normalized + [(0, None)]*back_pad
print("..  csv-table::")
print("    :header:Roll,Monster")
print()
for roll, (_, nm) in enumerate(ranks_padded, start=n_dice):
    if nm:
        print(f"    {roll},{monsters[nm].name}")
    else:
        print(f"    {roll},Reroll")
```

## Sub-Tables organized by realm

This will also use 3 dice, but, the dice will have easier-to-understand meanings.
There are 7 realms, which requires 2 dice.
Within a realm there are (currently) fewer than 6 inique creatures, so one die will work out nicely.

```python
def rank(thing):
    return (thing.budget['Attributes'] + thing.budget['Skills'] + thing.budget['Options']).n
```

```python
from collections import defaultdict
realms = defaultdict(list)
for name, thing in monsters.items():
    realms[thing.realm].append(thing)
realm_budget = [
    (sum(rank(thing) for thing in things), name)
    for name, things in realms.items()
]
realm_budget
```

While the above is helpful, it's still not exactly correct.

```python
total_realm = sum(budget for budget, name in realm_budget)
max_realm = max(budget for budget, name in realm_budget)
min_realm = min(budget for budget, name in realm_budget)
print(f"{total_realm=} {max_realm=}")
inverted_realm = [
    (max_realm-budget+min_realm, name)
    for budget, name in realm_budget
]
print(f"{inverted_realm=}")
total_inverted_realm = sum(budget for budget, name in inverted_realm)
print(f"{total_inverted_realm=}")
# TODO: Accumulate  rounding errors.
realm_rank = [
    (round(12*budget/total_inverted_realm), name,
     [m.name for m in monsters.values() if m.realm == name]
     )
    for budget, name in inverted_realm
]
```
