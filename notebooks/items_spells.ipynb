# Item and Enchantment Comparison

A Magical Item is a thing with an Enchantment bound to it. Most Items have charges. Rings (and Swords) are exceptions.

All the magical items in this world must have a corresponding spell.

We'll canvas the items and the spells by extracting the details from various notebooks.

```python
from pathlib import Path
import csv
import io
from itertools import chain

from IPython.display import display, Markdown, Latex

from opend6_tools.notebook_extract import *
```

```python
base = Path.cwd()
```

```python
def csv_table(heading, body_iter):
    buffer = io.StringIO()
    writer = csv.writer(buffer, delimiter="|")
    writer.writerow(heading)
    writer.writerow(["---"]*len(heading))
    writer.writerows(body_iter)
    framed = (
        f"|{line}|" for line in buffer.getvalue().splitlines()
    )
    return "\n".join(framed)
```

```python
def spells_table(spell_nb_iter):
    for nb in sorted(spell_nb_iter):
        spell_extractor = Extractor(nb, target_type=magic.Spell)
        spell_stmts = spell_extractor.definition_iter()
        spells = dict(
            eval_cell(name, stmt, EvalContext.MAGIC)
            for name, stmt in spell_stmts
        )
        for s in spells.values():
            yield nb.stem, s.name, s.effect.description()
```

```python
spell_nb_iter = chain(base.glob("* Workbook.ipynb"), base.glob("fantasy_Rank*.ipynb"))
display(
    Markdown(
        csv_table(
            ["file", "name", "description"],
            spells_table(spell_nb_iter)
        )
    )
)
```

```python
invocation_nb_iter = base.glob("fantasy_miracle_*.ipynb")
display(
    Markdown(
        csv_table(
            ["file", "name", "description"],
            spells_table(invocation_nb_iter)
        )
    )
)
```

```python
def items_table(item_nb_iter):
    for nb in sorted(item_nb_iter):
        item_extractor = Extractor(nb, target_type=magic.Spell)
        item_stmts = item_extractor.definition_iter()
        items = dict(
            eval_cell(name, stmt, EvalContext.MAGIC)
            for name, stmt in item_stmts
        )
        for item in items.values():
            yield nb.stem, item.name, item.effect.description()
```

```python
item_nb_iter = base.glob("unique_*.ipynb")
display(
    Markdown(
        csv_table(
            ["file", "name", "description"],
            items_table(item_nb_iter)
        )
    )
)
```
