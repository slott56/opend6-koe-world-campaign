# Heights and Weights

## Humans

| Gender | Measure     | Min | Mean | Max |
| ------ |-------------| --- | ---- | --- |
| Female | Mass (kg)   | 53  | 71   | 102 |
|        | Height (cm) | 154 | 163  | 172 |
| Male   | Mass (kg)   | 66  | 85   | 110 |
|        | Height (cm) | 167 | 177  | 187 |

```python
from dataclasses import dataclass

@dataclass
class Measure:
    unit: str
    min: int
    mean: int
    max: int

    @property
    def step(self) -> int:
        s = (self.max - self.min) / 15
        return int(s)

    @property
    def start(self) -> int:
        s = self.mean - 7.5 * self.step
        return int(s)

    @property
    def eqn(self) -> str:
        return f"{self.start:.0f} + {self.step:.0f} * roll"

    @property
    def min_c(self) -> int:
        return self.start + 3 * self.step
    @property
    def max_c(self) -> int:
        return self.start + 18 * self.step
    @property
    def mid_c(self) -> int:
        return self.start + 10 * self.step

female_mass = Measure("kg", 53, 71, 102)
female_height = Measure("cm", 154, 163, 172)

male_mass = Measure("kg", 66, 85, 110)
male_height = Measure("cm", 167, 177, 187)
```

```python
print("- Female")
print(f"  - Mass: {female_mass.eqn}")
print(f"    {female_mass.min_c} - {female_mass.mid_c} - {female_mass.max_c}")
print(f"  - Height: {female_height.eqn}")
print(f"    {female_height.min_c} - {female_height.mid_c} - {female_height.max_c}")

print("- Male")
print(f"  - Mass: {male_mass.eqn}")
print(f"    {male_mass.min_c} - {male_mass.mid_c} - {male_mass.max_c}")
print(f"  - Height: {male_height.eqn}")
print(f"    {male_height.min_c} - {male_height.mid_c} - {male_height.max_c}")
```
